```python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
X = iris['data'][:,3:]
print(iris.keys())
print(iris['feature_names'])
print(iris['target_names'])
print(X.shape)

binarized_target = np.where(iris['target'] == 2, 1, 0)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, binarized_target, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(max_iter=300)

model.fit(X_train, Y_train)

Y_prediction = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_prediction)
print(accuracy)
```

```python
import matplotlib.pyplot as plt

X_range = np.linspace(0, 3, 1000).reshape(-1, 1)
y_prob = model.predict_proba(X_range)
decision_boundary = X_range[y_prob[:, 1] >= 0.5][0]

plt.plot(X_range, y_prob[:, 1], "g-", label="Iris Virginica Probability")
plt.axvline(x=decision_boundary, color='red', linestyle='--', label='Decision boundary')
plt.title("Logistic regression - decision boundary")
plt.grid()
plt.legend()
plt.show()

print(f'결정 경계: {decision_boundary}')
```
